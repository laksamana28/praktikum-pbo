## Praktikum 1

```python
# Kelas Induk 
class Burung: 
    def __init__(self, nama): 
        self.nama = nama 

    def terbang(self): 
        print(f"{self.nama} terbang dengan cara umum.") 

    def bersuara(self): 
        print(f"{self.nama} mengeluarkan suara burung.")

# Kelas Anak 1 
class Elang(Burung): 
    def __init__(self, nama, rentang_sayap): 
        super().__init__(nama) 
        self.rentang_sayap = rentang_sayap 

    # Override 
    def terbang(self): 
        print(f"{self.nama} terbang tinggi melayang di angkasa.") 

    # Override 
    def bersuara(self): 
        print(f"{self.nama} berteriak nyaring!") 

# Kelas Anak 2 
class Pipit(Burung): 
    def __init__(self, nama, warna_bulu): 
        super().__init__(nama) 
        self.warna_bulu = warna_bulu 

    # Override 
    def terbang(self): 
        print(f"{self.nama} terbang cepat di antara pepohonan.") 

    # Override 
    def bersuara(self): 
        print(f"{self.nama} berkicau merdu: Cit cit!") 

# Fungsi yang memanfaatkan polimorfisme 
def demonstrasi_aksi_burung(daftar_burung): 
    print("\nAksi Burung:") 
    for burung in daftar_burung: 
        print(f"-- Aksi untuk {burung.nama} --") 
        # Panggilan metode yang sama, tapi perilaku berbeda tergantung objek 
        burung.terbang() 
        burung.bersuara() 
        print("-" * 15) 

# --- Kode Utama --- 
if __name__ == "__main__": 
    elang_jawa = Elang("Elang Jawa", 1.5) 
    pipit_gereja = Pipit("Pipit Gereja", "Coklat") 
    burung_aneh = Burung("Burung Misterius") # Objek dari kelas induk 

    koleksi_burung = [elang_jawa, pipit_gereja, burung_aneh] 

    demonstrasi_aksi_burung(koleksi_burung)
```

## Praktikum 2

```python
class Kucing: 
    def bersuara(self): 
        print("Kucing: Meow!") 

class Anjing: 
    def bersuara(self): 
        print("Anjing: Guk guk!") 

class Bebek: 
    def bersuara(self): 
        print("Bebek: Kwek kwek!") 

class Mobil: # Kelas ini TIDAK punya metode bersuara() 
    def jalankan(self): 
        print("Mobil: Brummm!") 

# Fungsi ini tidak peduli tipe objeknya, 
# asal punya metode .bersuara() 
def buat_suara(objek_yang_bisa_bersuara):
    try: 
        # Memanggil metode 'bersuara' jika ada 
        objek_yang_bisa_bersuara.bersuara() 
    except AttributeError: 
        # Menangani jika objek tidak punya metode 'bersuara'
        print(f"Objek {type(objek_yang_bisa_bersuara).__name__} tidak bisa bersuara.") 

# --- Kode Utama --- 
if __name__ == "__main__": 
    kucing1 = Kucing() 
    anjing1 = Anjing() 
    bebek1 = Bebek() 
    mobil1 = Mobil() 

    daftar_objek = [kucing1, anjing1, bebek1, mobil1] 

    print("Demonstrasi Duck Typing:") 
    for item in daftar_objek: 
        buat_suara(item)
```

## Praktikum 3

```python
class KalkulatorSederhana: 
    def __init__(self, nama="Kalkulator"): 
        self.nama = nama 
        print(f"{self.nama} siap digunakan.") 

    def tambah(self, *args): 
        """ 
        Metode ini menjumlahkan semua argumen angka yang diberikan. 
        Menerima sejumlah argumen posisi menggunakan *args. 
        """ 
        print(f"\nMemanggil metode tambah dengan argumen: {args}") 
        if not args: 
            print("Tidak ada angka untuk dijumlahkan.") 
            return 0 

        total = 0 
        valid_input = True 
        for angka in args: 
            if isinstance(angka, (int, float)): 
                total += angka 
            else: 
                print(f"Peringatan: Argumen '{angka}' bukan angka dan akan diabaikan.") 
                valid_input = False 

        if valid_input: 
            print(f"Hasil penjumlahan: {total}") 
        else: 
            print(f"Hasil penjumlahan (dengan beberapa input diabaikan): {total}") 
        return total 

# --- Kode Utama --- 
if __name__ == "__main__": 
    calc = KalkulatorSederhana("Calc-01") 

    print("\n--- Percobaan Penjumlahan ---") 
    calc.tambah(5, 10) 
    calc.tambah(2, 3, 5, 10) 
    calc.tambah(100) 
    calc.tambah() 
    calc.tambah(1, 2, "tiga", 4, 5.5)
```

## Praktikum 4

```python
class DataProcessor: 
    def __init__(self, processor_id): 
        self.processor_id = processor_id 
        print(f"Data Processor {self.processor_id} siap.") 

    def process(self, *args, **kwargs): 
        """ 
        Metode ini memproses data dengan argumen posisi (*args) 
        dan kata kunci (**kwargs) yang jumlahnya variabel. 
        """ 
        print(f"\n--- {self.processor_id} Memproses Data ---") 

        # Memproses argumen posisi (*args -> tuple) 
        if args: 
            print("Argumen Posisi Diterima:") 
            for i, arg in enumerate(args): 
                print(f" args[{i}]: {arg} (tipe: {type(arg).__name__})") 
        else: 
            print("Tidak ada argumen posisi.") 

        # Memproses argumen kata kunci (**kwargs -> dict) 
        if kwargs: 
            print("Argumen Kata Kunci Diterima:") 
            for kunci, nilai in kwargs.items(): 
                print(f" {kunci}: {nilai} (tipe: {type(nilai).__name__})") 
        else: 
            print("Tidak ada argumen kata kunci.") 
        print("---------------------------------") 

# --- Kode Utama --- 
if __name__ == "__main__": 
    # Membuat instance dari kelas 
    processor1 = DataProcessor("DP-001") 

    # Memanggil metode 'process' dengan berbagai cara 
    print("\nPanggilan 1: Tanpa argumen tambahan") 
    processor1.process() 

    print("\nPanggilan 2: Hanya argumen posisi") 
    processor1.process(100, "Status OK", 99.9, False) 

    print("\nPanggilan 3: Hanya argumen kata kunci") 
    processor1.process(user="admin", level=5, mode="verbose") 

    print("\nPanggilan 4: Kombinasi argumen posisi dan kata kunci") 
    processor1.process("Task-A", "Task-B", status="Running", priority="High", thread_id=54321) 

    # Panggilan dengan list dan dictionary sebagai argumen posisi 
    print("\nPanggilan 5: Argumen posisi kompleks") 
    list_ids = ["A1", "B2", "C3"] 
    dict_params = {"timeout": 30, "retry": 3} 
    processor1.process(list_ids, dict_params, owner="system", enabled=True)
```

```python
import math 

# --- Definisi Kelas dari Hierarki Inheritance --- 
class Bentuk: 
    def nama_bentuk(self): 
        return "Bentuk Generik" 

    def hitung_luas(self): 
        raise NotImplementedError("Subclass harus mengimplementasikan metode ini") 

class Lingkaran(Bentuk): 
    def __init__(self, radius): 
        self.radius = radius 

    def nama_bentuk(self): 
        return "Lingkaran" 

    def hitung_luas(self): 
        return math.pi * (self.radius ** 2) 

class Persegi(Bentuk): 
    def __init__(self, sisi): 
        self.sisi = sisi 

    def nama_bentuk(self): 
        return "Persegi" 

    def hitung_luas(self): 
        return self.sisi * self.sisi 

# --- Definisi Kelas Lain (Bekerja via Duck Typing) --- 
class TaplakMeja: 
    def __init__(self, panjang, lebar): 
        self.panjang = panjang 
        self.lebar = lebar 

    def hitung_luas(self): 
        return self.panjang * self.lebar 

    def deskripsi(self): 
        return f"Taplak Meja {self.panjang}x{self.lebar}" 

# --- Fungsi Polimorfik --- 
def tampilkan_info_luas(objek_dengan_luas): 
    print("-" * 20) 
    try: 
        luas = objek_dengan_luas.hitung_luas() 
        
        try: 
            nama = objek_dengan_luas.nama_bentuk() 
        except AttributeError: 
            nama = type(objek_dengan_luas).__name__ 
        
        print(f"Objek: {nama}") 
        print(f"Luas : {luas:.2f}") 
        
    except AttributeError: 
        print(f"Objek {type(objek_dengan_luas).__name__} tidak dapat dihitung luasnya.") 
    except NotImplementedError: 
        print(f"Metode 'hitung_luas' belum diimplementasi untuk {type(objek_dengan_luas).__name__}.") 

# --- Kode Utama --- 
if __name__ == "__main__": 
    lingkaran1 = Lingkaran(7) 
    persegi1 = Persegi(5) 
    taplak1 = TaplakMeja(1.5, 0.8) 
    bentuk_dasar = Bentuk() 
    string_biasa = "Ini string" 

    daftar_item = [lingkaran1, persegi1, taplak1, bentuk_dasar, string_biasa] 

    print("Menampilkan Info Luas (Polimorfisme Campuran):") 
    for item in daftar_item: 
        tampilkan_info_luas(item)
```

## Praktikum 6

```python
# --- Definisi Kelas Burung, Elang, Pipit --- 
class Burung: 
    def __init__(self, nama): 
        self.nama = nama 
    def terbang(self): 
        print(f"{self.nama} terbang dengan cara umum.") 
    def bersuara(self): 
        print(f"{self.nama} mengeluarkan suara burung.") 

class Elang(Burung): 
    def __init__(self, nama, rentang_sayap): 
        super().__init__(nama) 
        self.rentang_sayap = rentang_sayap 
    # Override 
    def terbang(self): 
        print(f"{self.nama} terbang tinggi melayang di angkasa.") 
    # Override 
    def bersuara(self): 
        print(f"{self.nama} berteriak nyaring!") 
    # Metode khusus Elang 
    def berburu(self): 
        print(f"{self.nama} sedang mencari mangsa dari ketinggian.") 

class Pipit(Burung): 
    def __init__(self, nama, warna_bulu): 
        super().__init__(nama) 
        self.warna_bulu = warna_bulu 
    # Override 
    def terbang(self): 
        print(f"{self.nama} terbang cepat di antara pepohonan.") 
    # Override 
    def bersuara(self): 
        print(f"{self.nama} berkicau merdu: Cit cit!") 
    # Metode khusus Pipit 
    def membangun_sarang(self): 
        print(f"{self.nama} sedang mengumpulkan ranting untuk sarang.") 

# --- Fungsi yang memanfaatkan polimorfisme DAN isinstance() --- 
def interaksi_dengan_burung(burung): 
    print(f"\n--- Berinteraksi dengan {type(burung).__name__}: {getattr(burung, 'nama', 'Objek tidak dikenal')} ---") 
    
    # Cek dulu apakah objek adalah instance dari Burung atau turunannya 
    if isinstance(burung, Burung): 
        # Perilaku polimorfik dasar (memanggil metode override) 
        burung.terbang() 
        burung.bersuara() 
        
        # Menambahkan perilaku spesifik berdasarkan tipe turunan 
        if isinstance(burung, Elang): 
            print("-> Ini adalah Elang!") 
            burung.berburu() 
        elif isinstance(burung, Pipit): 
            print("-> Ini adalah Pipit!") 
            burung.membangun_sarang() 
        else: 
            print("-> Ini adalah burung jenis umum (bukan Elang/Pipit).") 
    else: 
        print("-> Objek ini bukan termasuk jenis Burung.") 
    print("-" * 25) 

# --- Kode Utama --- 
if __name__ == "__main__": 
    elang_sumatra = Elang("Elang Sumatra", 1.8) 
    pipit_rumah = Pipit("Pipit Rumah", "Abu-abu") 
    merak = Burung("Merak") 
    kucing_tetangga = "Meong" 
    
    koleksi_makhluk = [elang_sumatra, pipit_rumah, merak, kucing_tetangga] 
    
    for makhluk in koleksi_makhluk: 
        interaksi_dengan_burung(makhluk)
```

```python
class Komputer:
    """Kelas induk yang merepresentasikan komputer secara umum."""

    def __init__(self, merk: str, processor: str, ram_gb: int):
        self.merk = merk
        self.processor = processor
        self.ram_gb = ram_gb

    def info_spesifikasi(self):
        """Mencetak informasi dasar komputer."""
        print(f"  Merk      : {self.merk}")
        print(f"  Processor : {self.processor}")
        print(f"  RAM       : {self.ram_gb} GB")

    def jalankan_aplikasi(self, nama_aplikasi: str):
        """Mencetak pesan bahwa komputer sedang menjalankan aplikasi."""
        print(f"  {self.merk} menjalankan aplikasi: {nama_aplikasi}...")

    # ── Simulasi Overloading – Opsi A (Default Argument) ──────────────────
    def upgrade_ram(self, tambahan_gb: int, tipe_ram: str = "DDR4"):
        """
        Upgrade RAM komputer.
        Simulasi overloading menggunakan default argument:
          - upgrade_ram(8)          → pakai default tipe_ram = "DDR4"
          - upgrade_ram(16, "DDR5") → tipe_ram ditentukan eksplisit
        """
        self.ram_gb += tambahan_gb
        print(
            f"  RAM {self.merk} diupgrade sebesar {tambahan_gb} GB "
            f"(Tipe: {tipe_ram}). Total RAM sekarang: {self.ram_gb} GB."
        )

    # ── Simulasi Overloading – Opsi B (*args) ─────────────────────────────
    def install_software(self, *args: str):
        """
        Install satu atau lebih software.
        Simulasi overloading menggunakan *args:
          - install_software("Chrome")
          - install_software("Chrome", "VLC", "VS Code")
        """
        for nama_software in args:
            print(f"  Menginstall {nama_software} di {self.merk}...")


# ─────────────────────────────────────────────
#  KELAS ANAK 1 – LAPTOP
# ─────────────────────────────────────────────
class Laptop(Komputer):
    """Kelas anak yang merepresentasikan laptop."""

    def __init__(
        self,
        merk: str,
        processor: str,
        ram_gb: int,
        ukuran_layar_inch: float,
        berat_kg: float,
    ):
        super().__init__(merk, processor, ram_gb)
        self.ukuran_layar_inch = ukuran_layar_inch
        self.berat_kg = berat_kg

    def info_spesifikasi(self):
        """Override: tampilkan spesifikasi komputer + info khusus laptop."""
        super().info_spesifikasi()
        print(f"  Ukuran Layar : {self.ukuran_layar_inch} inch")
        print(f"  Berat        : {self.berat_kg} Kg")


# ─────────────────────────────────────────────
#  KELAS ANAK 2 – DESKTOP
# ─────────────────────────────────────────────
class Desktop(Komputer):
    """Kelas anak yang merepresentasikan komputer desktop."""

    def __init__(
        self,
        merk: str,
        processor: str,
        ram_gb: int,
        jenis_casing: str,
        monitor_external: bool,
    ):
        super().__init__(merk, processor, ram_gb)
        self.jenis_casing = jenis_casing
        self.monitor_external = monitor_external

    def info_spesifikasi(self):
        """Override: tampilkan spesifikasi komputer + info khusus desktop."""
        super().info_spesifikasi()
        print(f"  Jenis Casing    : {self.jenis_casing}")
        status = "Ya" if self.monitor_external else "Tidak"
        print(f"  Monitor External: {status}")


# ─────────────────────────────────────────────
#  FUNGSI POLIMORFISME
# ─────────────────────────────────────────────
def cetak_semua_spesifikasi(daftar_komputer: list):
    """
    Demonstrasi polimorfisme:
    Fungsi memanggil info_spesifikasi() pada setiap objek dalam list.
    Metode yang sama menghasilkan output berbeda sesuai kelas masing-masing.
    """
    for i, komputer in enumerate(daftar_komputer, start=1):
        jenis = type(komputer).__name__
        print(f"\n{'─'*40}")
        print(f"  [{i}] {jenis.upper()}")
        print(f"{'─'*40}")
        komputer.info_spesifikasi()
    print(f"{'─'*40}\n")


# ─────────────────────────────────────────────
#  KODE UTAMA
# ─────────────────────────────────────────────
if __name__ == "__main__":

    # ── Buat objek Laptop ─────────────────────────────────────────────────
    laptop1 = Laptop(
        merk="ASUS VivoBook 15",
        processor="Intel Core i7-1255U",
        ram_gb=16,
        ukuran_layar_inch=15.6,
        berat_kg=1.8,
    )

    laptop2 = Laptop(
        merk="MacBook Air M3",
        processor="Apple M3",
        ram_gb=8,
        ukuran_layar_inch=13.6,
        berat_kg=1.24,
    )

    # ── Buat objek Desktop ────────────────────────────────────────────────
    desktop1 = Desktop(
        merk="Lenovo ThinkCentre",
        processor="Intel Core i9-13900K",
        ram_gb=32,
        jenis_casing="Tower",
        monitor_external=True,
    )

    desktop2 = Desktop(
        merk="Apple Mac Mini M4",
        processor="Apple M4",
        ram_gb=16,
        jenis_casing="Mini-PC",
        monitor_external=True,
    )

    # ── Masukkan ke dalam list ────────────────────────────────────────────
    daftar_semua_komputer = [laptop1, laptop2, desktop1, desktop2]

    # ── Demonstrasi Polimorfisme ──────────────────────────────────────────
    print("\n" + "=" * 40)
    print("   DEMONSTRASI POLIMORFISME")
    print("=" * 40)
    cetak_semua_spesifikasi(daftar_semua_komputer)

    # ── Demonstrasi jalankan_aplikasi ─────────────────────────────────────
    print("=" * 40)
    print("   DEMONSTRASI jalankan_aplikasi()")
    print("=" * 40)
    laptop1.jalankan_aplikasi("Microsoft Word")
    desktop1.jalankan_aplikasi("Adobe Premiere Pro")

    # ── Simulasi Overloading – Opsi A: upgrade_ram ───────────────────────
    print("\n" + "=" * 40)
    print("   SIMULASI OVERLOADING – Opsi A")
    print("   upgrade_ram (default argument)")
    print("=" * 40)

    # Tanpa argumen tipe_ram → pakai default "DDR4"
    print("\n[Tanpa argumen tipe_ram → default DDR4]")
    laptop1.upgrade_ram(8)

    # Dengan argumen tipe_ram eksplisit
    print("\n[Dengan argumen tipe_ram eksplisit → DDR5]")
    desktop1.upgrade_ram(32, "DDR5")

    # ── Simulasi Overloading – Opsi B: install_software ──────────────────
    print("\n" + "=" * 40)
    print("   SIMULASI OVERLOADING – Opsi B")
    print("   install_software (*args)")
    print("=" * 40)

    # Satu software
    print("\n[Satu software]")
    laptop2.install_software("Google Chrome")

    # Beberapa software
    print("\n[Beberapa software]")
    desktop2.install_software("VS Code", "Git", "Docker", "Postman")

    print("\n" + "=" * 40)
    print("   SELESAI")
    print("=" * 40 + "\n")
```
